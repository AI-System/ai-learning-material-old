```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# movie_data['popularity']
movie_top_20 = movie_data.sort_values(by='popularity').tail(20)[['original_title', 'popularity']]
# movie_top_20
ax = sb.barplot(x="original_title", y="popularity", data=movie_top_20)
plt.xticks(rotation=90)
```

**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# 新增：净利润列 = 票房 - 成本
movie_data['profit_my'] = movie_data['revenue'] - movie_data['budget']
new_data = movie_data.groupby(['release_year'])['profit_my']
new_data = new_data.agg('mean').reset_index().sort_values(by='release_year').tail(20)
# new_data
base_color = sb.color_palette()[2]
ax = sb.barplot(x="release_year", y="profit_my", data=new_data, color=base_color)
plt.xticks(rotation=90)
```
